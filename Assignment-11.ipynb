The PMF of X can be calculated using the binomial distribution:

P(X = k) = (20 choose k) * (11/44)^k * (33/44)^(20-k), for k = 0, 1, 2, ..., 20

This gives the probability that the student answers k questions correctly out of the 20 total questions.

To calculate P(X > 15), we need to sum the probabilities of X = 16, 17, 18, 19, and 20:

P(X > 15) = P(X = 16) + P(X = 17) + P(X = 18) + P(X = 19) + P(X = 20)

4. The number of students arriving at a college between a time interval is a Poisson random
variable. On average, 10 students arrive per hour. Let Y be the number of students arriving
from 10 am to 11:30 am. What is P(10&lt;Y≤15)?
    
Ans:
Since the number of students arriving is a Poisson random variable, we have:

P(Y = k) = (e^(-λ) * λ^k) / k!, for k = 0, 1, 2, ...

where λ is the mean number of students arriving per hour. In this case, λ = 10/60 * 1.5 = 0.25, since the time interval is 1.5 hours.

To find P(10 < Y ≤ 15), we can sum the probabilities of Y = 11, 12, 13, 14, and 15:

P(10 < Y ≤ 15) = P(Y = 11) + P(Y = 12) + P(Y = 13) + P(Y = 14) + P(Y = 15)

Using the Poisson distribution formula, we can calculate these probabilities as follows:

P(Y = 11) = (e^(-0.25) * 0.25^11) / 11! ≈ 0.0076

P(Y = 12) = (e^(-0.25) * 0.25^12) / 12! ≈ 0.0007

P(Y = 13) = (e^(-0.25) * 0.25^13) / 13! ≈ 0.00005

P(Y = 14) = (e^(-0.25) * 0.25^14) / 14! ≈ 0.000003

P(Y = 15) = (e^(-0.25) * 0.25^15) / 15! ≈ 0

5.Two independent random variables, X and Y,are given such that X~Poisson(α) and
Y~Poisson(β). State a new random variable as Z = X + Y. Find out the PMF of Z.

Ans:
Since X and Y are independent Poisson random variables with means α and β respectively, their probability mass functions are:

P(X = k) = (e^(-α) * α^k) / k!, for k = 0, 1, 2, ...
P(Y = k) = (e^(-β) * β^k) / k!, for k = 0, 1, 2, ...

To find the PMF of Z = X + Y, we need to calculate the probability that Z = k, for k = 0, 1, 2, ....

P(Z = k) = P(X + Y = k)

Since X and Y are independent, we can use the convolution formula to find the PMF of Z:

P(Z = k) = ∑(P(X = i) * P(Y = k - i)), for i = 0, 1, 2, ..., k

Substituting the Poisson PMFs, we get:

P(Z = k) = ∑((e^(-α) * α^i / i!) * (e^(-β) * β^(k-i) / (k-i)!)), for i = 0, 1, 2, ..., k

Simplifying the expression:

P(Z = k) = e^(-(α+β)) * ∑(α^i * β^(k-i) / (i! * (k-i)!)), for i = 0, 1, 2, ..., k

The sum in the expression above can be recognized as the binomial expansion of (α + β)^k. Therefore, we have:

P(Z = k) = e^(-(α+β)) * ((α + β)^k / k!), for k = 0, 1, 2, ...

This is the PMF of Z, which is a Poisson random variable with mean α + β.

8.There are 100 men on a ship. If Xi is the ith man&#39;s weight on the ship and Xi&#39;s are independent and
identically distributed and EXi = μ = 170 and σXi = σ = 30. Find the probability that the men&#39;s total
weight on the ship exceeds 18,000.

Ans:
Let X be the total weight of all 100 men on the ship. Since the Xi's are independent and identically distributed, X has a normal distribution with mean E[X] = E[ΣXi] = ΣE[Xi] = 100μ = 100 * 170 = 17000 and variance Var[X] = Var[ΣXi] = ΣVar[Xi] = 100σ^2 = 100 * 30^2 = 90000.

We want to find P(X > 18000). We can standardize X as Z = (X - E[X]) / sqrt(Var[X]) to obtain a standard normal distribution:

P(X > 18000) = P((X - E[X]) / sqrt(Var[X]) > (18000 - E[X]) / sqrt(Var[X]))

             = P(Z > (18000 - 17000) / 300)
             
             = P(Z > 1/3)
             
             = 1 - Φ(1/3)
             
             = 1 - 0.6293
             
             = 0.3707

Therefore, the probability that the men's total weight on the ship exceeds 18,000 is approximately 0.3707.
